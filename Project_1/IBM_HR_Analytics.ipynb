```python
# Import all the necessary python libraries

import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Import the csv file

data = pd.read_csv("/content/drive/MyDrive/WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
# Employee count by their Age

age = data[data['EmployeeCount']==1].groupby('Age').size()
age
```

```python
# Display Employee by their Age

plt.figure(figsize=(15,5))
sns.histplot(data=data,x='Age',kde=True,color='blue')
plt.title('Distribution of Employee by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#  Employee count by their Gender

gender = data[data['EmployeeCount']==1].groupby('Gender').size()
gender
```

```python
# Display Employee by their Gender

plt.figure(figsize=(5,5))
sns.histplot(data=data,x='Gender',color=('y'),alpha=0.7)
plt.title('Distribution of Employee by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#  Employee count by their Department

department = data[data['EmployeeCount']==1].groupby('Department').size()
department
```

```python
# Display Employee by their Department

plt.figure(figsize=(10,6))
sns.histplot(data=data,x='Department',color=('violet'))
plt.title('Distribution of Employee by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Display Attrition Count.

total_attrition = data['Attrition'].value_counts()
total_attrition
```

```python
# Create a Countplot for the attrition count.

plt.figure(figsize = (6,6))
sns.countplot(data = data, x= 'Attrition',color = 'green', edgecolor = 'black', alpha=0.7)
plt.title('Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
# Total employees count by department.

total_emp_dept = data[data['Attrition']== "Yes"].groupby('Department').size()
print(total_emp_dept)
```

```python
# Plot Countplot on count of employees by department.

plt.figure(figsize = (6,6))
sns.countplot(data = data, x= 'Department',color = 'navy', edgecolor = 'black', alpha=0.8)

plt.title('Employees Count by Department')
plt.xlabel('Department')
plt.ylabel('Count of employee')
plt.show()
```

```python
# Attrition count affected by distance from home.

attrition_Df = data[data['Attrition']== "Yes"].groupby('DistanceFromHome').size()
attrition_Df
```

```python
plt.figure(figsize=(10,6))
attrition_Df.plot(kind='area', alpha=0.4, stacked=True, color='red')
plt.title('Attrition Count by Distance from home')
plt.xlabel('Distance from home in KM')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(['No Attrition', 'Attrition'])
plt.tight_layout
plt.show()
```

```python
# Department wise attrition count.

attrition_dept = data[data['Attrition']== "Yes"].groupby('Department').size()
print(attrition_dept)
print(f"Total Attrition by department wise is :-", attrition_dept.sum())
```

```python
# Bar chart for attrition count on basis of department.

plt.figure(figsize=(6,6))
attrition_dept.plot(kind='bar', stacked = True, edgecolor='black', color=('pink'))
plt.title('Attrition Count by Department')
plt.xlabel('Department')
plt.ylabel('Count of employees')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Attrition count by overtime.

# Assuming you want to filter by the 'Attrition' column and count 'OverTime' values for those with 'Yes' attrition
attrition_overtime = data[data['Attrition']== "Yes"]['OverTime'].value_counts()
print(attrition_overtime)
print(f"Effect of overtime on Attrition Count :-", attrition_overtime.sum())
```

```python
# Pie Chart for attrition count by overtime.

plt.figure(figsize=(5,5))
plt.pie(attrition_overtime, labels=attrition_overtime, autopct='%1.1f%%', startangle=90,colors = sns.color_palette('deep', n_colors = len(attrition_overtime)))
plt.title('Attrition Count Affected by Overtime')
plt.legend(data['OverTime'], title='Overtime',loc='upper right')
plt.axis('equal')
plt.tight_layout()
```

```python
# Attrition count on basis of job satisfaction level.

attrition_JS = data[data['Attrition'] == "Yes"]['JobSatisfaction'].value_counts()
print(attrition_JS)
print(f"Attrition count on Job Satisfaction Basis :- ", attrition_JS.sum())
```

```python
# line chart for attrition on basis of job satisfaction level.

plt.figure(figsize=(8,5))
plt.plot(attrition_JS.index, attrition_JS.values, marker='o', markersize = '10', linestyle='-', color='y')

plt.title('Attrition Count on Job Satisfaction Level')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Attrition Count')
plt.xticks(attrition_JS.index)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Attrition count by years at company.

# Assuming you want to filter by the 'Attrition' column and count 'YearsAtCompany' values for those with 'Yes' attrition
attrition_yc = data[data['Attrition']== "Yes"]['YearsAtCompany'].value_counts()
print(attrition_yc)
print(f"Attrition count on 'Years at Company' Basis :-", attrition_yc.sum())
```

```python
# Bar chart for attrition count on basis of working years at company.

plt.figure(figsize=(10,8))
attrition_yc.plot(kind='bar', stacked = True,alpha = 0.7, edgecolor='black', color='orange')
plt.title('Attrition Count by No.of Years worked at the Company')
plt.xlabel('No.of years worked')
plt.ylabel('Attrition')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Attrition count on basis of Monthly Income.

attrition_MI = data.groupby(['MonthlyIncome','Attrition']).size().unstack(fill_value=0)
print(attrition_MI)
```

```python
# Scatter plot for attrition count on basis of monthly income.

plt.figure(figsize=(14,7))
sns.scatterplot(data=data,x='MonthlyIncome',y=data.index,hue='Attrition',style='Attrition',palette={'Yes': 'red', 'No': 'blue'},s=100)
plt.title('Attrition by Monthly Income')
plt.xlabel('Monthly Income')
plt.ylabel('Employee Index')
plt.xticks(rotation=0)
plt.grid=(True)
plt.legend(title='Attrition', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# Attrition count on basis of performance rating.

attrition_PR = data[data['Attrition'] == "Yes"]['PerformanceRating'].value_counts()
print(attrition_PR)
print(f"Attrition count on basis of Performance rating :-", attrition_PR.sum())
```
